```python
with open("train.txt", "a+") as f:
  img_list = os.listdir(train_img_path)
  for img in img_list:
    f.write(os.path.join(train_img_path,img+'\n'))
  print('Done')
```

```python
with open('train.txt', 'r') as file:
    contents = file.read()
    print(contents)
```
